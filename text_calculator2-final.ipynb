```python
def decoding(st):   #функция делит по словам и переводит слова в числа,  скидывает их в список

    dic1 = {"плюс": "+", "минус": "-", "умножить на": "*", "разделить на": "/", "делить на": "/", "в степени": "**", "скобка открывается": "(", "скобка закрывается": ")"}

    p = re.compile(r'( минус)|( плюс)|( умножить на)|( разделить на)|( делить на)|( в степени)|(скобка открывается)|( скобка закрывается)')
    lst = p.split(st)
    # print(lst)
    lst2 = []
    for el in lst:          #разделяет вводную строку на операции и числа 
        if el and el != ' ':
            el = re.sub(r'^ ', '', el)
            if el in dic1:
                lst2.append(dic1[el])
            elif el in dic:
                lst2.append(dic[el])
            else:
                el_split = el.split(' ')
                summ = 0
                for x in el_split:      #делить составные числа
                    # print(x)
                    summ += dic.get(x)
                lst2.append(summ)
    return lst2

# decoding('скобка открывается пять плюс два скобка закрывается умножить на скобка открывается три минус один')

```

```python
# Разворачиваем словать — меняем местами ключ значения
dict_num_words = dict([(num,word) for word,num in dic.items()])

def num_to_str(num):
    # Если выпадает ошибка, то её и возвращаем
    if num == 'Ошибка':
        return 'Ошибка'
    else:
        st = ''
        # Если число отрицательное, то добавляем слово минус в строку
        if num < 0:
            st += 'минус '
            num = -num
        if num < 20: # если число меньше 20, то подставляем его особенное слово
            st += dict_num_words[num]
        else:
            n = 0
            sp = []
            # раскладываем число на разряды, каждый новый разряд прибавляем к n единицу.
            while num > 0:
                ost = num % 10
                ost = ost * 10**n
                sp.append(ost) 
                n += 1
                num //= 10
            sp = sp[::-1] # разворачиваем список в обратном порядке

            # составляем фразу из чисел
            for elem in sp:
                if elem != 0:
                    st += dict_num_words[elem] + ' '
        return st
```
